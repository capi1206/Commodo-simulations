```python
import matplotlib.pyplot as plt
import numpy as np

def f(x, per, lag):
    return ((x+lag) % per)*np.where(x > lag, 1, 0)

def g(x):
    a=0
    for i in range(30):
        a =a+ f(x, 150 , i+5)
    return a    

# Generate x values
x_values = np.linspace(0, 370,3700)  # Creates 100 points between -5 and 5

y_values = g(x_values)

# Plot the graph
plt.plot(x_values, y_values, label='f(x) = x^2')
plt.title('Graph of f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

def f(x, per, lag, ext):
    return ((x - lag) % per) * np.where(x > lag, 1, 0) * np.where(x < lag + ext, 1, 0)

def g(x, periodo, extinguibilidad, num):
    a = 0
    for i in range(num):
        a = a + f(x, periodo, i, extinguibilidad)
    return a    

# Function to update the plot based on user parameters
def update_plot(periodo, extinguibilidad, num):
    x_values = np.linspace(0, 100, 1000)
    y_values = g(x_values, periodo, extinguibilidad, num)
    
    plt.figure(figsize=(8, 6))
    plt.plot(x_values, y_values, label='f(x)')
    plt.title('Total Commodos')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()

interact(update_plot, periodo=(1, 50, 1), extinguibilidad=(1, 50, 1), num=(1, 50, 1))
```
